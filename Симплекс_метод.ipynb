```python
import numpy as np
import math


e = 0.01 #Точность
n = 2 #Размерность задачи
m = 2 #Длина ребра симплекса
vec = np.array([1, 1]) #Переменная для записи координат с заданными начальными координатами
count = 0 #Счетчик итераций
function = np.array([]) #Значения функций

#Вычисляем приращение
q1 = ((math.sqrt(n+1) - 1)/(n*math.sqrt(2)))*m
q2 = ((math.sqrt(n+1) + n - 1)/(n*math.sqrt(2)))*m
print('Преращение:\n', 'Delta1 =', q1, '\n', 'Delta2 =', q2)

#Считаем координаты двух других вершин симплекса
x1 = np.array([vec[0] + q1, vec[1] + q2])
x2 = np.array([vec[0] + q2, vec[1] + q1])
vec = np.vstack([vec, x1, x2]) #Объединение(добавление)
print('Исходный базис:\n', vec)

def fun(x, y): #Подсчёт значения функции
    f = x**2 + 4*x*y + 17 * y**2 + 5 * y
    #f = 2.8 * y ** 2 + 1.9 * x + 2.7 * x ** 2 + 1.6 - 1.9 * y
    return f

def centrex(vect): #Подсчёт центра тяжести относительно других вершин
    xc = np.array([1/2*(vect[0][0] + vect[1][0]), 1/2*(vect[0][1] + vect[1][1])])
    return xc

def centrexs(vect):  #Подсчёт центра тяжести симплекса
    xc = np.array(1/3*vect[0] + 1/3*vect[1] + 1/3*vect[2])
    return xc

def regul(xci, xi): #Координаты отраженной вершины с помощью свойства регулярности
    xn = np.array(2*xci - xi)
    return xn

for i in range(3): #Добавление значение функции
    function = np.append(function, fun(vec[i][0], vec[i][1]))
f = function #Переменная для записи значиней функций базиса
vect = vec #Переменная для записи координат базиса
a = 0

while(True):
    print ('\nИтерация', count)
    k = np.argmax(f) #Номер вершины с наибольшим значением
    v = vect #Переменная нужна для подсчёта при редукции, поэтому приравниваем значению базиса
    vect = np.delete(vect,(k), axis = 0) #Удалить строку из координат наибольшего значения
    fu = f[k] #Наибольшее значение функции, нужно для сравнения
    func = f #Тоже нужно для редукции
    f = np.delete(f, (k), axis = 0) #Удалить строку из значений функции базиса
    xck = centrex(vect) #Считаем центр тяжести
    print("Центр тяжести:\n", xck)
    xo = regul(xck, v[k]) #Считаем отраженный вектор
    print("Отраженный вектор:\n", xo)
    fxo = fun(xo[0], xo[1]) #Новое значение функции(отраженная)
    if fxo < fu: #Если значение отраженной функции меньше наибольшего(прослеживается уменьшение значений)
        f = np.append(f, fxo) #Добавляем в базис
        vec = np.vstack([vec, xo])  # Добавить в список всех координат
        vect = np.vstack([vect, xo])  # Добавить в список базиса
        function = np.append(function, fxo)  # Добавить в спсок всех значений функций
    else: #Если функци увеличмлась, то проводим редукцию
        r = np.argmin(func)#Наименьшее значение целевой функции
        f = np.array(func[r]) #Базис обнуляется и присваивается первое из трех значений, которое является наименьшим из предыдущего базиса
        vect = v[r] #Тут тоже самое с координатами
        for j in range(3): #Цикл по предыдущему базису
            if func[j] != func[r]: #Если не равно наименьшему значению
                xnew = np.array(v[r] + 0.5 * (v[j] - v[r])) #Считаем новое значение координат. Значение v[j] берем из базиса,
                                                            # который брался в начале итерации. Было три значения в базисе, но берём два.
                                                            # Наименьшее не считается.
                                                            # К координатам наименьшего значения функции прибавляем 0.5 * на (значение координат из базиса - значение координат наименьшего значения функции)
                function = np.append(function, fun(xnew[0], xnew[1])) #Дальше просто заносим значения в базис и список всех значений
                f = np.append(f, fun(xnew[0], xnew[1]))
                vect = np.vstack([vect, xnew])
                vec = np.vstack([vec, xnew])
    xcs = centrexs(vect) #Считаем центр тяжести симплекса(координаты)
    fxs = fun(xcs[0], xcs[1]) #Считаем центр тяжести симплекса(функция)
    print("Векторное значение центра тяжести симплекса:\n", xcs)
    print("Численное значение центра тяжести симплекса:\n", fxs)
    print('Полученный базис:\n', vect, '\n')
    if (math.fabs(f[0] - fxs) < e and math.fabs(f[1] - fxs) < e and math.fabs(f[2] - fxs) < e):#Условие выхода
        break
    count += 1

a = np.argmin(f)

print('Таблица векторов:\n', vec, '\n') #Вывод без вектора, который не подошёл
print('Таблица значения функций:')
le = np.count_nonzero(function)
for i in range(le):
    print(' ', function[i])

print('Всего итераций:', count, '\n')
print('Минимальное значение функции:', f[a], 'при координатах:', vect[a])
```
